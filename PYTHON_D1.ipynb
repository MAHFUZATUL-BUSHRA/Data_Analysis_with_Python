```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(28,10)})

#plot
sns.countplot(df,x='housing_median_age',palette='rocket').set(
    title='Frequency of house by age',
  xlabel='age of house',
  ylabel='count of house'

)
plt.show()

```

```python
#get the pair value
plt.figure(figsize=(25,15))
#plot
sns.pairplot(df)
#show
plt.show()
```

```python
#Heatmap for observing correlation

heatmap=df.corr()

sns.heatmap(heatmap,annot=True, cmap='rocket').set(
    title="correlation between the data"
)
plt.xticks(rotation=360)
plt.show()
```

```python
#scatterplot(house price chnage with population)

sns.set(rc={'figure.figsize':(17,15)})

#plot
sns.scatterplot(data=df,x='population',y='median_house_value' , palette='rocket',alpha=0.4).set(
    title=" how House-price changes with population",
    xlabel='population',
    ylabel='House_value'
)
#plot
plt.show()
```

```python
#scatterplot(house price chnage with population)

sns.set(rc={'figure.figsize':(17,15)})

#plot
sns.kdeplot(data=df,x='population',y='median_house_value' ,fill='True' ,color='salmon').set(
    title=" Most common house price",
    xlabel='population',
    ylabel='House_value'
)
#plot
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,8)})

#plot
sns.boxplot(data=df,y='median_house_value', palette='rocket').set(
    title="house_value diversity",
    ylabel='House_value'
)
#plot
plt.show()
```
